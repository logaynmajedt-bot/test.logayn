```python
import pandas as pd; import matplotlib.pyplot as plt
print("Liberaries Imported Successfully")
```

```python
df = pd.read_csv("restaurant_orders.csv"); df.head()
```

```python
df.shape, df.dtypes
```

```python
df.describe(include="all")
```

```python
df.isna().sum()
```

```python
df["Discount"] = df["Discount"].fillna(0)
```

```python
df["Discount"] = df["Discount"] / 100
```

```python
print("Before conversion:", df["UnitPrice"].dtype)
df["UnitPrice"] = pd.to_numeric(df["UnitPrice"], errors="coerce")
print("After conversion:", df["UnitPrice"].dtype)
```

```python
print(df["UnitPrice"])
df["Discount"] = df["Discount"].fillna(0);df = df.dropna(subset=["UnitPrice"])
print(df["UnitPrice"])
```

```python
print(df["Date"])
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
print(df["Date"])
```

```python
print(df["City"])
df["City"] = df["City"].str.strip();
df = df.drop_duplicates(subset=["OrderID"])
print(df["City"])
```

```python
df["Revenue"] = df["Units"] * df["UnitPrice"] * (1 - df["Discount"])
print(df["Revenue"])
```

```python
df.dtypes
df.shape
```

```python
df["Revenue"].mean()
print(df["Revenue"].mean())
```

```python
df[["Units","UnitPrice","Discount","Revenue"]].describe()
```

```python
df["Revenue"].hist(bins=40);
plt.title(" ");
plt.xlabel("Revenue");
plt.ylabel("Count");
plt.show()
```

```python
df.groupby("Product")["Revenue"].sum().sort_values().plot.bar(); plt.title("   "); plt.show()
```

```python
df.groupby("Date")["Revenue"].sum().plot(); plt.title("   "); plt.xlabel("Date"); plt.ylabel("Revenue"); plt.show()
```
